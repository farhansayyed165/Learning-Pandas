```python
py_df.rename(columns = {'Country':'#Respindents','LanguageHaveWorkedWith':'num_knows_python'}, inplace=True)
```

```python
py_df['Pct_Knows_Python'] = (py_df['num_knows_python'] / py_df['#Respindents']) *100 
```
